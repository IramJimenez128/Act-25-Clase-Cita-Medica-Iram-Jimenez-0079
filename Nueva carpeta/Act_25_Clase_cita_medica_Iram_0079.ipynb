### Clase cita medica

```python
class CitaMedica:
    def __init__(self, producto, cantidad, ubicacion):
        # Atributos de la clase
        self.producto = producto
        self.cantidad = cantidad
        self.ubicacion = ubicacion

    # Funciones de la clase
    def leerdatos(self):
        print("\n--- Ingreso de Datos ---")
        self.producto = input("Ingrese el nombre del producto/servicio médico: ")
        self.cantidad = input("Ingrese la cantidad: ")
        self.ubicacion = input("Ingrese la ubicación de la cita: ")

    def mostrardatos(self):
        print("\n--- Detalles de la Cita Médica ---")
        print(f"Producto/Servicio: {self.producto}")
        print(f"Cantidad: {self.cantidad}")
        print(f"Ubicación: {self.ubicacion}")

    def cancelar_cita(self):
        # Tercera función necesaria para completar el requerimiento de 3 funciones
        print(f"\nLa cita de {self.producto} en {self.ubicacion} ha sido cancelada.")

# Creación de los objetos
cita1 = CitaMedica("Consulta General", "1", "Consultorio A")
cita2 = CitaMedica("Examen de Sangre", "3", "Laboratorio Central")

def menu():
    print("=== Sistema de Gestión de Citas ===")

    # Interactuando con cita1
    print("\n--- Gestionando Cita 1 ---")
    cita1.mostrardatos()
    cita1.leerdatos()
    cita1.mostrardatos()

    # Interactuando con cita2
    print("\n--- Gestionando Cita 2 ---")
    cita2.mostrardatos()
    cita2.leerdatos()
    cita2.mostrardatos

if __name__ == "__main__":
    menu()
print("Programa realizado por Iram Jimenez 0079 NL= 28")
```
